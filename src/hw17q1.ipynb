```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def entropy(f, n):
    ent = n * ((f-1)*np.log(1-f)-(f*np.log(f)))
    return ent
```

```python
n = np.power(10, 23)
nb = np.linspace(1, n-100, 1000000)
f_range = nb / n
print(f_range)
ent_range = entropy(f_range, n)
print(ent_range)
```

```python
fig = plt.figure()
plt.plot(f_range, ent_range)
plt.show()
```
