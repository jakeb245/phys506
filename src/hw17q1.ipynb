```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def entropy(f, n):
    ent = n * ((f-1)*np.log(1-f)-(f*np.log(f)))
    return ent
```

```python
n = np.power(10, 23)
nb = np.linspace(100, n-100, 100000)
f_range = nb / n
print(f_range)
ent_range = entropy(f_range, n)
print(ent_range)
```

```python
fig = plt.figure()
plt.plot(f_range, ent_range)
plt.title("Entropy of mixing molecules vs fraction f=Nb/N (N=10^23)", pad=15)
plt.xlabel("f")
plt.ylabel("Entropy")
plt.savefig("q1.png", dpi=300)
```
