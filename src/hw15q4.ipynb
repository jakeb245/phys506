The CNU parking deck can hold about 300 cars. They charge 1dol/day to park there. President Trible builds a brand-new parking deck which only has 100 parking spots. It costs 10dol/day to park there. If 200 people parked their cars randomly, make a graph of the fees collected vs. probability of collecting that fee. What's the mean, median, mode of this distribution?

```python
import random
import statistics
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_cars = 200
old_deck_cap = 300
old_deck_cost = 1
new_deck_cap = 100
new_deck_cost = 10
```

```python
n_runs = 10000

cost_array = []
for n in range(n_runs):
    cost = 0
    num_old = 0
    num_new = 0
    for i in range(n_cars):
        choice = random.randint(1,2)
        if choice == 2:
            if num_new >= new_deck_cap:
                num_old += 1
            else:
                num_new += 1
        else:
            num_old += 1

    cost = (num_old * old_deck_cost) + (num_new * new_deck_cost)
    cost_array.append(cost)
```

```python
cstart = min(cost_array)
cstop = max(cost_array)

prob = []

new_cost = []
for c in range(cstart, cstop+1):
    c_in_array = cost_array.count(c)
    if c_in_array > 0:
        prob.append(c_in_array / len(cost_array))
        new_cost.append(c)
```

```python
fig = plt.figure()
plt.plot(prob, new_cost)
plt.xlabel('Probability of collecting fee')
plt.ylabel('Parking fees collected ($)')
plt.title('Parking fees collected vs. Probability of collecting fee')
plt.show()
```

```python
mean = np.mean(cost_array)
median = np.median(cost_array)
mode = statistics.mode(cost_array)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```
